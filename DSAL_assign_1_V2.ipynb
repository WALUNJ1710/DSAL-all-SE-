```python
import random
import names
import pandas as pd
from csv import writer

tele_dict = {} #Blank Dictionary waiting for user to add records

#Declaration of functions...

def record_to_csv():              #Function to create CSV file to save record temporarily
    df = pd.DataFrame()
    df.to_csv("Telephone_Record.csv")


def add_tele_records(how_many):      #Function to add random telephone records
    with open('Telephone_Record.csv', 'w+', newline='') as writing_obj:
        labels = ['Name', 'Telephone Number']
        writer_obj = writer(writing_obj)
        writer_obj.writerow(labels)
        for i in range(how_many):
            client_name = names.get_first_name()
            client_last = names.get_last_name()
            clients = client_name + ' ' + client_last

            num1 = str(random.randint(7,9))
            num2 = str(random.randint(100000000, 999999999))
            client_num = num1+num2
            tele_dict.update({clients : client_num})

            client_list = [clients, client_num]
            writer_obj.writerow(client_list)

        writing_obj.close()
            
            

def search_record_byname(name):         #Function to search a record
    if name in tele_dict:
        print("{}'s telephone number is {}.\n".format(name, tele_dict[name]))
    else:
        print("No matching record found!\n")
        
def search_record_bynum(num):
    key_list = list(tele_dict.keys())
    val_list = list(tele_dict.values())
    if num in val_list:
        position = val_list.index(num)
        print("Name is {}.\n".format(key_list[position]))
    else:
        print("No matching record found!\n")
    
def add_tele_manual(client_name, client_number):    #Function to add records to telephone manually
    tele_dict.update({client_name:client_number})
    
#Start of code...

exit = 0
record_to_csv()
greet = ['Welcome to smart Tele-Book!', 'Welcome to SmartBook!', 'Welcome to Tele-Book!']
print(random.choice(greet),"\nWhat would you like to do?\n")

while exit != 1:
    choice = int(input("1 = Insert record to telephone book\n2 = Search record in telephone book\n3 = View records\n4 = Exit\nChoice = "))
    if choice == 1:
        chc2 = int(input("1 = Insert manually\n2 = Auto insert 'n' records\nChoice = "))
        if chc2 == 1:
            get_name = input("Enter name: ")
            get_num = input("Enter number: ")
            add_tele_manual(get_name, get_num)

        elif chc2 == 2:
            num_ = int(input("How many records do you want to add? "))
            add_tele_records(num_)
            print("Record added successfully!\n")

        else:
            print("Invalid Choice!\n")

    elif choice == 2:
        chc3 = int(input("1 = Search by name\n2 = Search by number\nChoice = "))
        if chc3 == 1:
            name_ = input("Enter name to find in record : ")
            search_record_byname(name_)
        
        elif chc3 == 2:
            num_ = int(input("Enter number to find in record : "))
            search_record_bynum(str(num_))
            
        else:
            print("Invalid Choice!\n")

    elif choice == 3:
        if len(tele_dict) == 0:
            print("No previous records found!\n")
        else:
            df = pd.read_csv('Telephone_Record.csv')
            print(df)

    elif choice == 4:
        exit += 1

    else:
        print("Invalid Choice!\n")
```
